To use `tidypandas` in a project:

```python
import tidypandas
import numpy as np

print(tidypandas.__version__)
```

```python
from tidypandas import tidyframe
from tidypandas.tidy_utils import simplify
from palmerpenguins import load_penguins
import pandas.api.types as dtypes
penguins_tidy = tidyframe(load_penguins())
penguins_tidy
```

```python
# penguins_tidy.mutate({'blmm': lambda x: x['bill_length_mm'] + 1}, by = 'species')

# penguins_tidy.group_modify(lambda x: x.assign(blmm = lambda x: x['bill_length_mm'] + 1)
#                            , by = 'species'
#                            , is_pandas_udf = True
#                           )

(penguins_tidy.to_pandas()
              .assign(__rn = lambda x: np.arange(x.shape[0]))
              .groupby('sex', dropna = False)
              .apply(lambda x: x.assign(blmm = lambda x: x['bill_length_mm'] + 1))
              .drop(columns = "__rn")
              .pipe(simplify)
#               .reset_index(drop = True)
#               .sort_values(by = "__rn")
#               .drop(columns = "__rn")
              )
```
